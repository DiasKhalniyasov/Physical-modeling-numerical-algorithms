# Tridiagonal matrix parabolic type implementation from scratch

```python
import math
import matplotlib.pyplot as plt
n= 10
x= 0.0001
A= 1/ x**2
B= -2/ x**2
C= 1/ x**2
D= 1
f=[]
for i in range(n):
    f+=[i/n]
f1=[]
for i in range(n):
    f1+=[i/n]
alpha=[0]*n
beta=[0]*n
beta[1]=1
alpha[1]=0
for i in range(1,n):
    if i<n-1:
        alpha[i+1] = -A/(C*alpha[i]+B)
        beta[i+1] = (D-C*beta[i])/(C*alpha[i]+B)
    else:
        pass

print(alpha)
print(beta)
u=[0]*n
u[-1]= (beta[-1])/(1-alpha[-1])

for i in range(n-1,0,-1):
    u[i-1] = alpha[i]*u[i]+beta[i]

analytic=[]
for i in range(n):
    analytic.append([1 - x**2*i*(n-1) + (x**2*i**2)/2])
print(u)
plt.plot(f, u, 'b', f1, analytic, 'k')
plt.show()
```
