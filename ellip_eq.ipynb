```python
import matplotlib.pyplot as plt
import numpy as np

n = 10
dx = 0.001
A = 1/dx**2
B = -2/dx**2
C = A
D = 1
xs = []
y = []
alfa = [0]*n
beta = [0]*n
u = [0]*n

for i in range(2,n):
    alfa[1] = 0
    beta[1] = 1
    alfa[i] = -A/(C*alfa[i-1]+B)
    beta[i] = (D-C*beta[i-1])/(C*alfa[i-1]+B)

u[n-1] = beta[n-1]/(1-alfa[n-1])

for i in range(2,n+1):
    u[n-i] = (beta[n-i+1]+u[n-i+1]*alfa[n-i+1])
print(u)

x = 0
for i in range(n):
    x = i*dx
    xs += [x]
    y += [1 - (n-1)*dx*x + x*x/2]

fig = plt.figure()
fig, ax = plt.subplots()
ax.plot(xs, u) #kok
ax.plot(xs, y) #orange
plt.show()
```
